```python
#processes the files properly from the NetLogo output
def process_nc_files(df1):
    df = df1.dropna(subset=['model-time-step-damage-list'])
    df['model-time-step-damage-list'] = df['model-time-step-damage-list'].str.replace(']', "") 
    df['model-time-step-damage-list'] = df['model-time-step-damage-list'].str.replace('[', "")
    df['model-time-step-damage-list'] = df['model-time-step-damage-list'].str.split(" ")
    df['model-time-step-damage-list'] = df['model-time-step-damage-list'].apply(lambda x : pd.to_numeric(x)) 
    df['regional_time_step_damage'] = [sum(a) if isinstance(a, (list, tuple, np.ndarray)) else a for a in  df['model-time-step-damage-list'].tolist()]
    df['regional_time_step_damage-billions'] = df['regional_time_step_damage'] / 1000000000
    df.rename(columns={'sea-level-rise': 'Sea Level Rise','mun-slr-proj': "Agents' Perceived Sea Level Rise Projection", "model-version": "Adaptation Coordination"}, inplace=True)
    
    df['Adaptation Coordination'] = df['Adaptation Coordination'].replace("no cooperation", "No Cooperation")
    df['Adaptation Coordination'] = df['Adaptation Coordination'].replace("regional authority", "Regional Authority")
    df['Adaptation Coordination'] = df['Adaptation Coordination'].replace("voluntary cooperation", "Voluntary Cooperation")
    
    df['model-cumulative-adaptation-cost-list'] = df['model-cumulative-adaptation-cost-list'].str.replace(']', "") 
    df['model-cumulative-adaptation-cost-list'] = df['model-cumulative-adaptation-cost-list'].str.replace('[', "")
    df['model-cumulative-adaptation-cost-list'] = df['model-cumulative-adaptation-cost-list'].str.split(" ")
    #df['Features'] = df['Features'].apply(lambda x : pd.to_numeric(x))
    df['model-cumulative-adaptation-cost-list'] = df['model-cumulative-adaptation-cost-list'].apply(lambda x : pd.to_numeric(x)) 
    df['model-cumulative-adaptation-costs'] = [sum(a) if isinstance(a, (list, tuple, np.ndarray)) else a for a in df['model-cumulative-adaptation-cost-list'].tolist()]                                                                            
    df['model-cumulative-adaptation-costs-billions'] = df['model-cumulative-adaptation-costs'] / 1000000000
    
    df['model-cumulative-damage-list'] = df['model-cumulative-damage-list'].str.replace(']', "") 
    df['model-cumulative-damage-list'] = df['model-cumulative-damage-list'].str.replace('[', "")
    df['model-cumulative-damage-list'] = df['model-cumulative-damage-list'].str.split(" ")
    #df['Features'] = df['Features'].apply(lambda x : pd.to_numeric(x))
    df['model-cumulative-damage-list'] = df['model-cumulative-damage-list'].apply(lambda x : pd.to_numeric(x)) 
    df['regional_cumulative_damage'] = [sum(a) if isinstance(a, (list, tuple, np.ndarray)) else a for a in df['model-cumulative-damage-list'].tolist()]
    df['regional_cumulative_damage-billions'] = df['regional_cumulative_damage'] / 1000000000

    
    df['model-adaptation-list'] = df['model-adaptation-list'].str.replace(']', "") 
    df['model-adaptation-list'] = df['model-adaptation-list'].str.replace('[', "")
    df['model-adaptation-list'] = df['model-adaptation-list'].str.replace('none ', "")
    df['model-adaptation-list'] = df['model-adaptation-list'].str.replace(' none', "")
    df['model-adaptation-list'] = df['model-adaptation-list'].str.replace('none', "")
    df['model-adaptation-list'] = df['model-adaptation-list'].str.split(" ")

    df.loc[df['Adaptation Coordination'] == 'No Cooperation', 'Adaptation Costs'] = df['model-cumulative-adaptation-costs']
    df.loc[df['Adaptation Coordination'] == 'No Cooperation', 'Harbor Damages'] = df['regional_time_step_damage']
    df.loc[df['Adaptation Coordination'] == 'Voluntary Cooperation', 'Adaptation Costs'] = df['model-cumulative-adaptation-costs']
    df.loc[df['Adaptation Coordination'] == 'Voluntary Cooperation', 'Harbor Damages'] = df['regional_time_step_damage']
    df.loc[df['Adaptation Coordination'] == 'No Cooperation', 'Cumulative Harbor Damages'] = df['regional_cumulative_damage']
    df.loc[df['Adaptation Coordination'] == 'Voluntary Cooperation', 'Cumulative Harbor Damages'] = df['regional_cumulative_damage']

    print(df["Harbor Damages"].head())
    print(df["Cumulative Harbor Damages"].head())
    return df

import numpy as np
nc_full_processed = process_nc_files(nc_full)


```
